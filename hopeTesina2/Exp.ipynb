```python
import numpy as np
import matplotlib.pyplot as plt

# Read the data from pso_log.txt
with open('pso_log.txt', 'r') as file:
    lines = file.readlines()

# Parse the data into a list of lists
data = []
labels = []
for line in lines:
    label, values = line.split(':')
    labels.append(label.strip())
    data.append([float(value.split('(')[1].split(')')[0]) for value in values.split(',')])

# Plot all the graphs in the same figure with labels
for graph, label in zip(data, labels):
    plt.plot(graph, label=label)

plt.title('Best over Iteration')
plt.xlabel('Iteration')
plt.ylabel('Fitness Score')
plt.legend()
```
